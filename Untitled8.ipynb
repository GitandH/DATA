```python
import matplotlib.pyplot as plt
import pandas as pd


plt.rc('font',family='NanumBarunGothic')

plt.rcParams['axes.unicode_minus'] = False
```

```python
df=pd.read_csv('/content/행정구역별인구현황.csv')
df
```

```python
df['행정구역'] = df['행정구역'].str(:-12)
df
```

```python
df.columns = ['행정구역','총인구수','새대수','새대당인구','남자인구수','여자인구수','남녀비율']
df
```

```python
df = df.drop(index=0)
df
```

```python
df.nlargest(5,'총인구수')
```

```python
import seaborn as sns
```

```python
plt.figure(figsize=(15,5))
plt.bar(df['행정구역'],df['총인구수'],color = 'red')
plt.xticks(rotation=45)
plt.title('행정구역별 총인구수')
plt.figure(figsize=(15,5))
plt.show()
```

```python
plt.figure(figsize=(15,5))
df_t=df.sort_values(by='새대당인구',ascending=False)
plt.plot(df_t['행정구역'],df_t['새대당인구'],color='red',marker='o')
plt.xticks(rotation=45)
plt.title('가구당 인구수')
plt.grid(visible=True)
plt.show
```
