```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df_temp = pd.read_csv('/content/seoul_temp.csv')
df_temp.head(5)
df_temp
```

```python
df_temp.tail(10)
```

```python
# df_temp.set_index('연도',inplace=True)
df_temp
```

```python
plt.plot(df_temp.index,df_temp['평균기온'])
plt.show()
```

```python
plt.plot(df_temp.index,df_temp['평균기온'])
plt.plot(df_temp.index,df_temp['최고기온'])
plt.title('한글')
plt.show()
```

```python
plt.figure(figsize=(20,10))
plt.plot(df_temp.index,df_temp['평균기온'],color = 'red',marker='>')
plt.plot(df_temp.index,df_temp['최고기온'],color = 'orange',marker='<')
plt.plot(df_temp.index,df_temp['최저기온'],color = 'cyan',marker='s')
plt.title('서울시연간기온')
plt.xlabel('연도')
plt.ylabel('기온')
plt.grid()
plt.show()
```

```python
label = ['A','B','AB','O','ABC']
size = [34.8,32.9,14.7,10,7.6]
fig,ax = plt.subplots(ncols=3,nrows=2, figsize=(6,6))
ax[0,1].plot(np.arange(0,11,1))
ax[1,2].plot(np.arange(11,0,-1))
ax[0,0].scatter(np.random.rand(50),np.random.rand(50))
ax[0,2].pie(size,labels=label,autopct='%.1f%%')
ax[1,1].bar(df_temp.index,df_temp['평균기온'],color = 'r')
ax[1,0].pie(size,labels=label,autopct='%.1f%%')
plt.show
```
