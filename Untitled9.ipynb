```python
import matplotlib.pyplot as plt
plt.rc('font',family='NanumBarunGothic')

plt.rcParams['axes.unicode_minus'] = False
```

```python
import pandas as pd

df_eat = pd.read_csv('/content/청소년식생활현황.csv')
df_exe = pd.read_csv('/content/청소년신체운동현황.csv')
df_fat = pd.read_csv('/content/청소년비만.csv')
df_mind = pd.read_csv('/content/청소년정신건강.csv')
df_smoke = pd.read_csv('/content/청소년흡연율현황.csv')
df_drink = pd.read_csv('/content/청소년음주현황.csv')
```

```python
df_merge=df_eat[['시점','전체아침결식','전체주3회정크푸트']].copy()
df_merge = pd.merge(df_merge,df_fat[['시점','전체비만율','전체이미지왜곡인지']],on = '시점',how='left')
df_merge
```

```python
plt.bar(df_merge['시점'],df_merge['전체비만율'],label='비만율',color = 'r')
plt.plot(df_merge['시점'],df_merge['전체아침결식'],marker = 'o',color='b',label='아침결식')

plt.legend()
plt.grid()
plt.title('비만율과 아침결식비율관계')
plt.xlabel('연도')
plt.ylabel('비율')
plt.show()
```

```python
df_merge= pd.merge(df_merge, df_drink[['시점','전체음주율']],on = '시점',how='left')
df_merge=pd.merge(df_merge,df_smoke[['시점','전체흡연율']], on ='시점',how='left')
df_merge
```

```python
df_merge
```

```python
df_merge.plot(kind='bar',x='시점',y=['전체음주율_y','전체흡연율'],color=['b','r'],width=0.5,figsize=(10,5))
plt.title('청소년 흡연,음주행태')
plt.ylabel('단위(%)')
plt.show()
```

```python
df_mind
```

```python
df_mind.plot(kind='area',x='시점',y=['남학생우울증','여학생우울증'],color=['b','r'],alpha = 0.2)
plt.title('우울감경험률')
plt.ylabel('단위(%)')
plt.show()
```
